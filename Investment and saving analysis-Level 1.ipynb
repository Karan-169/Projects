```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

### **Step- 2 Load the Dataset**

```python
df = pd.read_csv('Investment and saving data.csv')
df.head()
```

### **Step - 3 Performing The Descriptive analysis**

```python
df.info()
```

#### **Identifying the null value** 

```python
df.isnull().sum().sum()
```

#### **Identifying the duplicate value**

```python
df.duplicated().sum()
```

#### **Identify summary of statistics for numerical or categorical columns**

```python
df.describe()
```

### **Step-5 Identify and Vizualise the gender distribution** 

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = 'gender', data=df)

ax.bar_label(ax.containers[0])
plt.title("gender distribution")
plt.show()
```
